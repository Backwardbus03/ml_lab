```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.info()
```

```python
for col in df.columns:
  if df[col].isnull().any():
    if df[col].isna().sum() >= 500:
      df.drop(columns=col, inplace=True)
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.info()
```

```python
numeric_data = df.select_dtypes(include=[np.number])

corr = numeric_data.corr()["SalePrice"]

for col in corr.index:
  if abs(corr[col]) < 0.3:
    df.drop(columns=col, inplace=True)

df.info()
```

```python
y = df['SalePrice']
x = df.drop(columns='SalePrice', axis=1)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

df['SalePrice_cat'] = pd.cut(df['SalePrice'],
                               bins=[0, 100000, 150000, 200000, 250000, 300000, 400000, 500000, np.inf],
                               labels=[1, 2, 3, 4, 5, 6, 7, 8])

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(x, df['SalePrice_cat']):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]


df.drop('SalePrice_cat', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

x_train = x.iloc[train_index]
x_test = x.iloc[test_index]


categorical_cols = x_train.select_dtypes(include='object').columns


x_train_encoded = pd.get_dummies(x_train, columns=categorical_cols, drop_first=True)


x_test_encoded = pd.get_dummies(x_test, columns=categorical_cols, drop_first=True)

x_train_final, x_test_final = x_train_encoded.align(x_test_encoded, join='left', axis=1, fill_value=0)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train_final)
x_test = scaler.transform(x_test_final)


```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

reg = RandomForestRegressor()

param = {
    "n_estimators": [10, 100, 200, 500, 1000],
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10]
}

grid_search = GridSearchCV(reg, param, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(x_train, y_train)
```

```python
final_clf = grid_search.best_estimator_
final_clf
```

```python
y_pred = final_clf.predict(x_test)
```
